# Hypothesis Test 1 - MPAA Rating


## Hypothesis

Null Hypothesis - There is no correlation between MPAA rating and revenue.

Alternative Hypothesis - There are some correlations between any/all ratings and revenue.

```python
sns.barplot(data = df, x='certification', y = 'revenue')
```

## ANOVA and assumptions

For this test I've decided to go with an ANOVA test

### Defining Groups

```python
needed_cols = ['certification', 'revenue']
df[needed_cols]
```

```python
groups = {}
## Loop through all unique categories
for i in df['certification'].unique():
    ## Get series for group and rename
    data = df.loc[df['certification']==i,'revenue']
    
    # save into the dictionary
    groups[i] = data
groups.keys()
```

### Assumption 1 - Removing Outliers

```python
temp = np.abs(stats.zscore(groups['PG'])) > 3
temp.value_counts()
```

```python
for rating, data in groups.items():
    outliers = np.abs(stats.zscore(data)) > 3
    print(f"there were {outliers.sum()} outliers in the {rating} group")
    data = data.loc[~outliers]
    groups[rating] = data
```

### Assumption 2 - Normality

```python
norm_results = {}
for i, data in groups.items():
    stat, p = stats.normaltest(data)
    ## save the p val, test statistic, and the size of the group
    norm_results[i] = {'n': len(data),
                             'p':p,
                             'test stat':stat,}
## convert to a dataframe
norm_results_df = pd.DataFrame(norm_results).T
norm_results_df
```

```python
norm_results_df['sig'] = norm_results_df['p'] < .05 
norm_results_df
```

#### Result

Test showed no significant normality, however the samples are large enough to ignore this assumption.

### Assumption 3 - Equal Variance

```python
result = stats.levene(*groups.values())
print(result)
```

```python
if result.pvalue < .05:
    print("the groups do not have equal variance")
else:
    print("the groups do have equal variance")
```

### Perform Test

Test selected is the kruskal test since we violated the third assumption of variance.

```python
result = stats.kruskal(*groups.values())
result
```

```python
sns.barplot(data = df, x='certification', y = 'revenue')
```

## Reject the Null Hypothesis

With results of a p value lower than an Alpha of 0.05 we may reject the null hypothesis that there is no correlation between movie rating and revenue. 

From the visualization above we can see that pg and pg13 movies are among the highest revenues produced by each category.

## Hypothesis Test 3 - Long movies v.s Short movies

### Hypothesis

Null Hypothesis -  Movie length does not correlate to increased or decreased revenue

Alternative Hypothesis - Length of movie correlates to increased or decreased revenue.

### Get Groups

```python
df.head()
```

```python
df['runtime'].isna().sum()
```

```python
df['revenue'].isna().sum()
```

```python
sns.barplot(data = df, x = 'runtime', y = 'revenue')
```

```python
long_movies = df.loc[df['runtime'] > 150, 'revenue']
short_movies = df.loc[df['runtime'] < 150, 'revenue']
```

```python
print(long_movies.value_counts().sum())
print(short_movies.value_counts().sum())
```

### Remove Outliers

```python
zscores_l= stats.zscore(long_movies)
outliers_l = abs(zscores_l)>3
np.sum(outliers_l)
```

```python
long_movies = long_movies[(np.abs(stats.zscore(long_movies)) < 3)]
```

```python
zscores_s= stats.zscore(short_movies)
outliers_s = abs(zscores_s)>3
np.sum(outliers_s)
```

```python
short_movies = short_movies[(np.abs(stats.zscore(short_movies)) < 3)]
```

### Check for normality

```python
result_l = stats.normaltest(long_movies)
result_l
```

```python
result_s = stats.normaltest(short_movies)
result_s
```

#### normality

Though we do violate normality we can safely ignore due to sample size

### Equal Variance

```python
result = stats.levene(long_movies, short_movies)
result
```

#### variance
We do not have equal variance and will use equal_var = False on t-test 

```python
result = stats.ttest_ind(long_movies, short_movies, equal_var = False)
result
```

## Reject the Null Hypothesis

With a p-value less than 0.05 we can safely reject the null hypothesis.

We can assume there is correlation between runtime and revenue.
